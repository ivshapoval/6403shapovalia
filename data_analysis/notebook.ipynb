```python
from datetime import datetime
from timerow.module.functions import WeatherData
from meteostat import Point, Daily
import pandas as pd
```

```python
# Получение данных через meteostat

start = datetime(2020, 1, 1)
end = datetime(2024, 11, 7)

# Точка для измерения погоды - Самара
samara = Point(53.2001, 50.15, 116)

# Получение данных о погоде в Самаре
data = Daily(samara, start, end)
data = data.fetch()
```

```python
# Инициализация модуля для расчета данных о погоде

weather = WeatherData(data)
weather.get_data()
```

```python
# Вычисление скользящего среднего 
weather.moving_average(21)
# Вычисление разницы между значениями скользящего среднего
weather.differentiate(21)
```

```python
# Поиск экстремальных значений скользящего среднего
weather.find_extremal(21)
```

```python
# Поиск экстремальных значений скользящего среднего
weather.autocorrelation(21)
```

```python
# Сохранение данных в results.xlsx
if data is not None:
    data.to_excel("results.xlsx")
    print(f"Данные успешно сохранены в results.xlsx")
else:
    print("Нет данных для сохранения.")
```
